### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Load data

```python
d = 3
```

```python
val_mc = np.load('results/sw_d'+str(d)+'.npy')
val_cv = np.load('results/swnn_d'+str(d)+'.npy')
I_true = np.load('results/I_true_d'+str(d)+'.npy')

val_mc = val_mc-I_true
val_cv = val_cv-I_true
```

### Display boxplots of SW-MC vs. SW-CVNN

```python
fig, ax = plt.subplots(figsize=(5,5))
bp_mc = ax.boxplot(val_mc.T,
                   positions=2*np.arange(5)-0.3,
                   sym='',patch_artist=True)
bp_cv = ax.boxplot(val_cv.T,
                   positions=2*np.arange(5)+0.3,
                   sym='',patch_artist=True)
for patch in bp_mc['boxes']:
    patch.set_alpha(0.7)
    patch.set_facecolor('blue')
    
for patch in bp_mc['medians']:
    patch.set_color('cyan')
    
for patch in bp_cv['boxes']:
    patch.set_alpha(0.7)
    patch.set_facecolor('green')
    
for patch in bp_cv['medians']:
    patch.set_color('lime')

plt.plot([], c='blue', label='SW-MC')
plt.plot([], c='green', label='SW-CVNN')
plt.xticks(ticks=2*np.arange(5),
           labels=['50','100','250','500','1000'],
           fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Number of Random Projections',fontsize=15)
plt.ylabel('Error distribution',fontsize=15)
plt.legend(fontsize=15)
plt.grid(alpha=0.3,linestyle='--')

plt.show()
```
